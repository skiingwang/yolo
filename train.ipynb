```python
import time, numpy as np, paddle, paddle.nn as nn, matplotlib.pyplot as plt, os
from paddle.io import DataLoader
from yolov1 import Yolo1, YoloLoss
from dataset import YoloDataset, collate_fn, preprocessor

device = 'cuda' if paddle.cuda.is_available() else 'cpu'
```

```python
model = Yolo1()
paddle.summary(model, (1, 3, 448, 448))
```

```python
transform = paddle.vision.transforms.Compose([
    paddle.vision.transforms.Resize((448, 448)),
    # paddle.vision.transforms.Normalize(mean=[0.4914, 0.4822, 0.4465], std=[0.2470, 0.2435, 0.2616])
    paddle.vision.transforms.ToTensor()
])

ds = YoloDataset('./data/Drink_284_Detection', transform=transform)
train_loader = DataLoader(ds, batch_size=3, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(ds, batch_size=3, shuffle=False, collate_fn=collate_fn)

d = next(iter(train_loader))
len(d[2])
```

```python
def train(model, save_path, train_loader, val_loader, epochs=10, lr=0.0001, s=7, b=2, c=20):
    optimizer = paddle.optimizer.AdamW(parameters=model.parameters(), learning_rate=lr)
    loss_fn = YoloLoss()
    losses = []
    val_losses = []
    since = time.time()
    for epoch in range(epochs):
        model = model.to(device).train()
        print(f'Epoch {epoch+1}/{epochs}')
        batch_losses = []
        for batch_idx, batch in enumerate(train_loader):
            img, labels, boxes = batch
            img = img.to(device)
            logits = model(img)  # [batch_size, s, s, b * 5 + c]
            target = preprocessor(labels, boxes, s, b, c, logits)
            loss = loss_fn(logits, target)
            batch_losses.append(loss)
            optimizer.clear_grad()
            loss.backward()
            optimizer.step()
        losses.append(np.mean(batch_losses))
        print(f'Loss: {losses[-1]:.4f}')
        print('-'*32)
    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed//60:.0f}m {time_elapsed%60:.0f}s')
    return losses, val_losses
```

```python
train(model, './yolo1.pd', train_loader, val_loader)
```
