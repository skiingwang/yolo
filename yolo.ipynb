```python
import paddle, paddle.nn.functional as F, numpy as np
```

```python
# 卷积层+BN层
class ConvBNLayer(paddle.nn.Layer):
    def __init__(self, ch_in, ch_out, kernel_size=3, stride=1, groups=1, padding=0, act='leaky'):
        super().__init__()
        # 卷积层
        self.conv = paddle.nn.Conv2D(
            in_channels=ch_in,
            out_channels=ch_out,
            kernel_size=kernel_size,
            stride=stride,
            padding=padding,
            groups=groups,  # 分组卷积
            weight_attr=paddle.ParamAttr(initializer=paddle.nn.initializer.Normal(0., 0.02)), bias_attr=False)  # 权重初始化为正态分布，不使用偏置项
        # BN层，对卷积输出进行归一化，加速收敛并防止过拟合
        self.batch_norm = paddle.nn.BatchNorm2D(
            num_features=ch_out,
            # 归一化包括gamma和beta参数，用于对特征图进行缩放和偏移，gamma需要正则化(L2范数正则)，则beta不需要
            weight_attr=paddle.ParamAttr(initializer=paddle.nn.initializer.Normal(0., 0.02),  regularizer=paddle.regularizer.L2Decay(0.0001)),
            bias_attr=paddle.ParamAttr(initializer=paddle.nn.initializer.Constant(0.0),  regularizer=paddle.regularizer.L2Decay(0.)))
        self.act = act

    def forward(self, inputs):  # 卷积->批归一->激活
        out = self.conv(inputs)
        out = self.batch_norm(out)
        if self.act == 'leaky':
            out = F.leaky_relu(x=out, negative_slope=0.1)
        return out

# 下采样层
class DownSample(paddle.nn.Layer):
    # 直接调用ConvBNLayer，将步长设置为2，图片尺寸减半
    def __init__(self, ch_in, ch_out, kernel_size=3, stride=2, padding=1):
        super().__init__()
        self.conv_bn_layer = ConvBNLayer(ch_in=ch_in, ch_out=ch_out, kernel_size=kernel_size, stride=stride, padding=padding)

    def forward(self, inputs):
        return self.conv_bn_layer(inputs)

# 基本残差块
class BasicBlock(paddle.nn.Layer):
    # 输入x经过两层卷积，然后将第二层卷积的输出和输入x相加
    def __init__(self, ch_in, ch_out):
        super().__init__()
        # 第一个卷积层用于降维，将通道数减半
        self.conv1 = ConvBNLayer(ch_in=ch_in, ch_out=ch_out, kernel_size=1, stride=1, padding=0)
        # 第二个卷积层用于特征提取，将通道数加倍
        self.conv2 = ConvBNLayer(ch_in=ch_out, ch_out=ch_out*2, kernel_size=3, stride=1, padding=1)

    def forward(self, inputs):
        conv1 = self.conv1(inputs)
        conv2 = self.conv2(conv1)
        # 将第二层卷积的输出和输入x相加，实现残差连接
        out = paddle.add(x=inputs, y=conv2)
        return out

# 残差块层级，将多个BasicBlock堆叠在一起
class LayerWarp(paddle.nn.Layer):
    # count代表残差块数量，组成Darknet53网络的一个层级
    def __init__(self, ch_in, ch_out, count, is_test=True):
        super().__init__()
        # 第一个BasicBlock，处理输入
        self.basicblock0 = BasicBlock(ch_in, ch_out)
        self.res_out_list = []
        for i in range(1, count):
            # 使用add_sublayer添加子层
            res_out = self.add_sublayer("basic_block_%d" % (i), BasicBlock(ch_out*2, ch_out))
            self.res_out_list.append(res_out)

    def forward(self,inputs):
        y = self.basicblock0(inputs)
        for basic_block_i in self.res_out_list:
            y = basic_block_i(y)
        return y

# DarkNet主体，来自DarkNet的网络结构图
DarkNet_cfg = {53: [1, 2, 8, 8, 4]}

class DarkNet53_conv_body(paddle.nn.Layer):
    def __init__(self):
        super().__init__()
        self.stages = DarkNet_cfg[53]
        self.stages = self.stages[:]
        # 初始输入通道数为3，输出通道数为32
        self.conv0 = ConvBNLayer(ch_in=3, ch_out=32, kernel_size=3, stride=1, padding=1)
        # 第一个下采样层，步幅为2，将特征图尺寸减半
        self.downsample0 = DownSample(ch_in=32, ch_out=32 * 2)

        # 添加各个层级的实现
        self.darknet53_conv_block_list = []  # 残差块层
        self.downsample_list = []  # 下采样层
        for i, stage in enumerate(self.stages):
            # 根据DarkNet_cfg构建残差块层级
            conv_block = self.add_sublayer('stage_%d' % (i), LayerWarp(32*(2**(i+1)), 32*(2**i), stage))
            self.darknet53_conv_block_list.append(conv_block)
        # 两个层级之间使用DownSample将尺寸减半
        for i in range(len(self.stages) - 1):
            downsample = self.add_sublayer('stage_%d_downsample' % i, DownSample(ch_in=32*(2**(i+1)), ch_out=32*(2**(i+2))))
            self.downsample_list.append(downsample)

    def forward(self,inputs):
        out = self.conv0(inputs)
        # print("conv1:",out.numpy())
        out = self.downsample0(out)
        # print("dy:",out.numpy())
        blocks = []  # 存储每个层级的输出
        # 依次将各个层级作用在输入上面
        for i, conv_block_i in enumerate(self.darknet53_conv_block_list):
            out = conv_block_i(out)
            blocks.append(out)
            if i < len(self.stages) - 1:
                out = self.downsample_list[i](out)
        return blocks[-1:-4:-1] # [start, stop, step]，将C0, C1, C2作为返回值，对应于大、中、小特征图
```

```python
'''DarkNet53 网络结构
BackBone（DarkNet53）=（残差块+下采样）*（2+3【C0, C1, C2】）
残差块 = 降维卷积 + 特征提取卷积 + 残差连接
卷积块 = 卷积层 + 批归一化层 + LeakyReLU激活函数
'''
backbone = DarkNet53_conv_body()
x = paddle.to_tensor(np.random.randn(1, 3, 640, 640).astype('float32'))
C0, C1, C2 = backbone(x)
print(f'C0 Shape: {C0.shape}\nC1 Shape: {C1.shape}\nC2 Shape: {C2.shape}')
```
